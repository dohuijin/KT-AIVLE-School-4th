```python
import mido
import random

# 미디 파일을 저장할 경로와 파일 이름

filename = 'test.mid'
copy_filename='/content/base_music_basic_pitch.mid'
# 미디 트랙 생성
mid = mido.MidiFile()
track = mido.MidiTrack()
mid.tracks.append(track)

midi_data = load_midi(copy_filename)

mean_note=0
note_num=0
for tracks in midi_data.tracks:
  for mids in tracks:
    if mids.type=='note_on':
      mean_note+=mids.note
      note_num+=1
mean_note//=note_num
mean_note
all_time=0
prev_mid=0
pitch_size=2.4
is_start=True
minus_note=7
for tracks in midi_data.tracks:
  for ori_mid in tracks:
    if ori_mid.type=='note_on':
      if ori_mid.velocity==0 or ori_mid.note-minus_note<=59 or ori_mid.note-minus_note>=75:
        all_time+=int(ori_mid.time*pitch_size)
      else:
        if is_start:
          is_start=False
          prev_mid=ori_mid
          note_on = mido.Message('note_on', note=ori_mid.note-minus_note, velocity=100, time=int(ori_mid.time*pitch_size))
          track.append(note_on)
        else:
          note_off = mido.Message('note_off', note=prev_mid.note-minus_note, velocity=100, time=all_time)
          track.append(note_off)
          all_time=0
          prev_mid=ori_mid
          note_on = mido.Message('note_on', note=ori_mid.note-minus_note, velocity=100, time=int(ori_mid.time*pitch_size))
          track.append(note_on)
    else:
      all_time+=int(ori_mid.time*pitch_size)
note_off = mido.Message('note_off', note=prev_mid.note-minus_note, velocity=100, time=all_time+100)
track.append(note_off)

# 미디 파일 저장
mid.save('/content/mlp-singer/data/raw/mid/'+filename)
```

```python
from pydub import AudioSegment
from pydub.playback import play

def mix_music(background_file, foreground_file, output_file):
    # 배경음악 파일 로드
    background_music = AudioSegment.from_file(background_file)

    # 전경음악 파일 로드
    foreground_music = AudioSegment.from_file(foreground_file)

    # 두 음악 파일의 길이 맞추기
    if len(background_music) < len(foreground_music):
        background_music = background_music._spawn(
            b"\0" * (len(foreground_music) - len(background_music)) + background_music.raw_data
        )
    else:
        foreground_music = foreground_music._spawn(
            b"\0" * (len(background_music) - len(foreground_music)) + foreground_music.raw_data
        )

    # 두 음악 파일 병렬로 섞기
    mixed_music = background_music.overlay(foreground_music)

    # 결과를 새로운 파일로 저장
    mixed_music.export(output_file, format="wav")

if __name__ == "__main__":
    background_file = "/content/base_music.wav"  # 배경음악 파일
    foreground_file = "/content/mlp-singer/samples/test_generated_e2e.wav"  # 전경음악 파일
    output_file = "/content/output_mixed_music1.wav"  # 섞인 출력 파일

    # 두 음악 파일 병렬로 섞기
    mix_music(background_file, foreground_file, output_file)

```
