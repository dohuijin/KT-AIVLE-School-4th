# [Step 2] 데이터 전처리

#### **<span style="color:blue">[2-1] air_21, air_22 의 '측정일시'를 활용하여 'time' 변수 생성</span>**
* air_21, air_22  각각 '측정일시'를 활용하여 'time'변수 생성
    * 참고: 미세먼지 데이터는 1시-24시, 날씨 데이터는 0시-23시로 구성되어 있습니다.

```python
# air_21, air_22 각각 '측정일시'를 활용하여 'time'변수 생성
# time 변수를 pd.to_datetime으로 데이터 타입 변경

air_21['time'] = air_21['측정일시'] - 1
air_21['time'] = pd.to_datetime(air_21['time'], format='%Y%m%d%H')

air_21['time']
```

```python
air_22['time'] = air_22['측정일시'] - 1
air_22['time'] = pd.to_datetime(air_22['time'], format='%Y%m%d%H')

air_22['time']
```

#### **<span style="color:blue">[2-2] weather_21, weather_22 의 '일시'를 활용하여 'time' 변수 생성</span>**
* weather_21, weather_22 의 '일시'를 활용하여 'time'변수 생성
* time 변수를 pd.to_datetime으로 데이터 타입 변경

```python
weather_21['time'] = weather_21['일시']
weather_21['time'] = pd.to_datetime(weather_21['time'])

# 'weather_22'
weather_22['time'] = weather_22['일시']
weather_22['time'] = pd.to_datetime(weather_22['time'])
weather_21['time'].head()
```

#### **<span style="color:blue">[2-3] 'time' 기준으로 데이터 합치기</span>**
* 미세먼지 데이터와 날씨 데이터를 'time' 기준으로 합쳐보세요.
    * df_21에는 'time' 기준으로 21년도 미세먼지, 날씨 데이터를 합쳐보세요.
    * df_22에는 'time' 기준으로 22년도 미세먼지, 날씨 데이터를 합쳐보세요.

```python
df_21 = pd.merge(air_21, weather_21, on='time', how='inner')

df_22 = pd.merge(air_22, weather_22, on='time', how='inner')

df_21
```

```python
df_21.info()
```

#### **<span style="color:blue">[2-4] 사용하지 않을 변수 제거</span>**
* 머신러닝에 사용하지 않을 변수들을 제거

```python
# df_21, df_22에 사용할 변수들만 할당
selected_columns = ['time', 'PM10', 'PM25','SO2','CO','O3','NO2', '기온(°C)', '강수량(mm)','풍속(m/s)','습도(%)']
df_21 = df_21[selected_columns]
df_22 = df_22[selected_columns]
```

```python
# time 변수를 index로 세팅
df_21 = df_21.set_index('time').sort_index()
df_22 = df_22.set_index('time').sort_index()

df_21
```

```python
df_21.isnull().sum()
```

```python
df_22.isnull().sum()
```

```python
# df_21, df_22의 변수 중 결측치를 처리
df_21.drop(['강수량(mm)'], axis=1, inplace=True)
df_22.drop(['강수량(mm)'], axis=1, inplace=True)
```

```python
# 나머지 전날 값으로 결측치 채우기
df_21.fillna(method='ffill', inplace=True)
df_22.fillna(method='ffill', inplace=True)
```

```python
# df_21, df_22의 결측치 재확인
print(df_21.isnull().sum())
print(df_22.isnull().sum())
```

#### **<span style="color:blue">[2-6] 전일 같은 시간 미세먼지 농도 변수 추가</span>**

```python
# month, day, hour 변수를 추가
df_21['month'] = df_21.index.month
df_21['day'] = df_21.index.day
df_21['hour'] = df_21.index.hour

df_22['month'] = df_22.index.month
df_22['day'] = df_22.index.day
df_22['hour'] = df_22.index.hour
```

```python
#df_21, df_22에 전일 같은 시간 미세먼지 농도 변수(PM10_lag1) 추가

df_21['PM10_lag1'] = df_21['PM10'].shift(24)

df_22['PM10_lag1'] = df_22['PM10'].shift(24)

df_21
```

#### **<span style="color:blue">[2-7] t+1 시점의 미세먼지 농도 데이터 생성</span>**
* t+1 시점은 1시간 후 입니다.
* t+1 시점의 미세먼지 농도 변수(PM10_1)를 생성.
* t+1 시점의 미세먼지 농도는 머신러닝 모델을 통해 예측하려는 y값(target)







```python
# df_21, df_22에 t+1 시점 변수(PM10_1) 추가
df_21['PM10_1'] = df_21['PM10'].shift(-1)
df_22['PM10_1'] = df_22['PM10'].shift(-1)
```

```python
# 결측치 이전 값으로 대체
df_21['PM10_1'].fillna(method='ffill', inplace=True)

df_22['PM10_1'].fillna(method='ffill', inplace=True)

df_21['PM10_lag1'].fillna(df_21['PM10'], inplace=True)

df_22['PM10_lag1'].fillna(df_22['PM10'], inplace=True)
```

```python
df_21.isnull().sum()
```

#### **<span style="color:blue">[2-8] train, test 데이터 분리</span>**

```python
train_x = df_21.drop(columns=['PM10_1'])
train_y = df_21['PM10_1']

test_x = df_22.drop(columns=['PM10_1'])
test_y = df_22['PM10_1']
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
print(train_x.info())
print(test_x.info())
```

#[3-0] 머신러닝 모델링

```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
import joblib
```

**[3-1]모델링: LinearRegression**
* 성능지표 : MSE, R-squared Score(소수점 5째자리까지)

```python
model1 = LinearRegression()
model1.fit(train_x,train_y)
```

```python
pred1 = model1.predict(test_x)
```

```python
mse1 = mean_squared_error(test_y, pred1)
r2_1 = r2_score(test_y, pred1)

print(f'MSE: {mse1:.5f}')
print(f'R-squared Score: {r2_1:.5f}')
```

```python
# 학습 성능
print('train score :', model1.score(train_x, train_y))

# 평가 성능
print('test score :', model1.score(test_x, test_y))
```
